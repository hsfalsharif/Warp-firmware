```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
from scipy.fft import fft

matplotlib.use('TkAgg')
```

```python
df1 = pd.read_csv('CW4_data/backwards_forwards_handwave1.csv')
df2 = pd.read_csv('CW4_data/backwards_forwards_handwave2.csv')
df3 = pd.read_csv('CW4_data/backwards_forwards_handwave3.csv')
df4 = pd.read_csv('CW4_data/backwards_forwards_handwave4.csv')
df5 = pd.read_csv('CW4_data/backwards_forwards_handwave5.csv')

df6 = pd.read_csv('CW4_data/backwards_forwards_handwave6.csv')
df7 = pd.read_csv('CW4_data/backwards_forwards_handwave7.csv')
df8 = pd.read_csv('CW4_data/backwards_forwards_handwave8.csv')
df9 = pd.read_csv('CW4_data/backwards_forwards_handwave9.csv')
df10 = pd.read_csv('CW4_data/backwards_forwards_handwave10.csv')#

df11 = pd.read_csv('CW4_data/backwards_forwards_handwave11.csv')
df12 = pd.read_csv('CW4_data/backwards_forwards_handwave12.csv')
df13 = pd.read_csv('CW4_data/backwards_forwards_handwave13.csv')
df14 = pd.read_csv('CW4_data/backwards_forwards_handwave14.csv')
df15 = pd.read_csv('CW4_data/backwards_forwards_handwave15.csv')
```

```python
x_spacing = np.linspace(0, 501, num=501)

# average out signals
dataframes = np.array([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12, df13, df14, df15])
#dataframes = np.array([df1, df2, df3, df4, df5])
combined_df_avg = np.zeros((501, 3))
combined_df_samples = np.zeros((501, 3, 15))
for df_idx in range(len(dataframes)):
    df_x = dataframes[df_idx][:, 0]
    df_y = dataframes[df_idx][:, 1]
    df_z = dataframes[df_idx][:, 2]
    for i in range(len(df_x)):
        combined_df_avg[i][0] += df_x[i]
        combined_df_avg[i][1] += df_y[i]
        combined_df_avg[i][2] += df_z[i]

        combined_df_samples[i][0][df_idx] = df_x[i]
        combined_df_samples[i][1][df_idx] = df_y[i]
        combined_df_samples[i][2][df_idx] = df_z[i]

combined_df_avg = combined_df_avg / 15
combined_df_stddev = np.zeros_like(combined_df_avg, dtype=np.float64)

for i in range(len(combined_df_stddev)):
    for axis in range(len(combined_df_stddev[i])):
        combined_df_stddev[i][axis] = np.std(combined_df_samples[i][axis])

combined_df_avg = combined_df_avg.astype(int)
combined_df_stddev = combined_df_stddev.astype(int)
        
```
